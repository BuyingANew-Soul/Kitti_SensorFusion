```python
class DatasetHandler():

    def __init__(self, sequence, lidar=True, low_memory=True):
        self.lidar = lidar
        self.low_memory = low_memory
        self.seq_dir = f"../dataset/gray/sequences/{sequence}/"
        # self.velodyne_dir = f"../dataset/velodyne/sequences/{sequence}/velodyne"
        self.pose_dir = f"../dataset/poses/{sequence}.txt"
        poses = pd.read_csv(self.pose_dir, delimiter= ' ', header=None)
        
        self.left_image_files = os.listdir(self.seq_dir+"/image_0/")
        print(self.left_image_files[0])
        self.right_image_files = os.listdir(self.seq_dir+"/image_1/")

        self.lidar_path =  f"../dataset/velodyne/sequences/{sequence}/velodyne/"
        self.velodyne_files = os.listdir(self.lidar_path)
        
        self.num_frames = len(self.left_image_files)
        

        poses = pd.read_csv("../dataset/poses/00.txt", delimiter = " ", header=None)
        self.ground_truth_poses = np.zeros((len(poses), 3, 4))
        for i in range(len(poses)):
            self.ground_truth_poses[i] = np.array(poses.iloc[i]).reshape(3,4)

        calib = pd.read_csv(f"../dataset/calibration/sequences/{sequence}/calib.txt", delimiter = " ", header=None, index_col = 0)
        self.P0 = np.array(calib.loc['P0:']).reshape(3,4)
        self.P1 = np.array(calib.loc['P1:']).reshape(3,4)
        self.P2 = np.array(calib.loc['P2:']).reshape(3,4)
        self.P3 = np.array(calib.loc['P3:']).reshape(3,4)
        self.Tr = np.array(calib.loc['Tr:']).reshape(3,4)
        

        if low_memory:
            self.reset_frames()
            print(self.seq_dir+ '/image_0/' + self.left_image_files[0])
            self.first_image_left = cv2.imread(self.seq_dir+ '/image_0/' + self.left_image_files[0], 0)
            self.second_image_left = cv2.imread(self.seq_dir+ '/image_0/' + self.left_image_files[0], 0)
            self.first_image_right = cv2.imread(self.seq_dir+ '/image_1/' + self.right_image_files[0], 0)
            self.second_image_right = cv2.imread(self.seq_dir+ '/image_1/' + self.right_image_files[0], 0)
            self.first_point_cloud = np.fromfile(self.lidar_path + self.velodyne_files[0], dtype=np.float32).reshape(-1,4)

            self.image_height = self.first_image_left.shape[0]
            self.image_width = self.first_image_left.shape[1]

    
    def reset_frames(self):
        self.left_images = (cv2.imread(self.seq_dir + '/image_0/' + img, 0) for img in self.left_image_files)
        self.right_images = (cv2.imread(self.seq_dir + '/image_1/' +  img, 0) for img in self.right_image_files)
        self.point_clouds = (np.fromfile(self.lidar_path + point_cloud, dtype=np.float32).reshape(-1,4) 
                             for point_cloud in self.velodyne_files)


    def show_image(self):
        %matplotlib widget
        next_image = next(self.left_images)
        print(next_image)
        plt.imshow(next_image)
        
```

```python
handler = DatasetHandler('00')
```

```python
handler.show_image()
```
